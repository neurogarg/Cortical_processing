#### __Question 1:__ Looking at the files, I found that electrophysiological data is present in only those files which have the word behaviour in it. Does this mean only for these sessions recordings were done and can I safely disregard other files?  

```python
import h5py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

data_file = r"/Volumes/Elements/succ_rep/tongue/000239"

files = [f for f in os.listdir(data_file) if os.path.isdir(os.path.join(data_file, f))]

## choosing only those files which have behaviour in the name
session_file = []
for f in files:
    subject =  os.path.join(data_file, f)
    sub_sess = os.listdir(subject)
    for file in sub_sess:
        if "_behavior" in file:
            session_file.append(os.path.join(subject, file))
#session_file
```

```python
hf = h5py.File(session_file[0], 'r')
keys = hf.keys()
```

```python
trials_df = {}
for key in hf["intervals"]["trials"].keys():
    if hf["intervals"]["trials"][key][:].shape[0] == 265:
        trials_df[key] = hf["intervals"]["trials"][key][:]

df = pd.DataFrame(trials_df)
df
```

```python
## adding lick on times and lick off times based on the index values
counter_matlab_idx = 1
lick_On_times = hf["intervals"]["trials"]["lickOn_times"][:]
df ["lick_On_times"] = lick_On_times[df["lickOn_times_index"]-counter_matlab_idx]

lick_Off_times = hf["intervals"]["trials"]["lickOff_times"][:]
df ["lick_Off_times"] = lick_Off_times[df["lickOff_times_index"]-counter_matlab_idx]

df
```

```python
## adding position times based on the index
df["posIndex_times"] = [[] for _ in range(len(df))]


pos_time_indices = list(df["posIndex_times_index"])
pos_time_indices.insert(0, 0)
pos_time_index_pairs = list(zip(pos_time_indices, pos_time_indices[1:]))

for i, (start, end) in enumerate(pos_time_index_pairs):
    pos_index_times_array = np.array(hf["intervals"]["trials"]["posIndex_times"][start:end])
    df.at[i, "posIndex_times"] = pos_index_times_array.tolist()

df
```
